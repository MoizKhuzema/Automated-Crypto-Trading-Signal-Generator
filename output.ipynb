## **REAL TIME DATA COLLECTION**

DATA: HISTORICAL PRICE DATA FOR CRYPTO COINS

API : BINANCE API

```python
import numpy as np
from binance.client import Client
```

```python
# Binance API key
BINANCE_API_KEY = ''
BINANCE_API_SECRET = ''
```

```python
# Create Binance client
client = Client(BINANCE_API_KEY, BINANCE_API_SECRET, requests_params={"timeout": 30})

# Fetch all trading symbols available on Binance
exchange_info = client.get_exchange_info()
symbols = [symbol['symbol'] for symbol in exchange_info['symbols']]
print(f'Number of symbols: {len(symbols)}')
print(f'First 10 symbols: {symbols[:10]}')
```

```python
def fetch_crypto_data(symbol, interval, limit=10):
    # Fetch real-time data from Binance API
    klines = client.get_klines(symbol=symbol, interval=interval, limit = limit)
    timestamps = np.array([kline[0] // 1000 for kline in klines], dtype= 'double')  # Convert milliseconds to seconds
    open_prices = np.array([float(kline[1]) for kline in klines], dtype= 'double')
    high_prices = np.array([float(kline[2]) for kline in klines], dtype= 'double')
    low_prices = np.array([float(kline[3]) for kline in klines], dtype= 'double')
    close_prices = np.array([float(kline[4]) for kline in klines], dtype= 'double')
    volumes = np.array([float(kline[5]) for kline in klines], dtype= 'double')
    return timestamps, open_prices, high_prices, low_prices, close_prices, volumes
```

```python
# Set interval for 1 hour
interval = Client.KLINE_INTERVAL_1HOUR
output = []

# Fetching real-time data for COIN ETHBTC
timestamp, open_prices, high_prices, low_prices, close_prices, volumes = fetch_crypto_data(symbols[0], interval)

print(f"""
COIN: {symbols[0]}

time stamp: {timestamp}

open prices: {open_prices}

high prices: {high_prices}

close prices: {close_prices}

volume: {volumes}
""")
```

```python
def predict(template, coin, time_frame, high_price, low_price, close_price, volume, risk_tolerance, minimum_profit, indicator_data):
    # Initialize model
    llm = OpenAI(openai_api_key=OPENAI_API_KEY, model_name=OPENAI_MODEL, temperature=0.9)

    # generate prompt (input)
    prompt = PromptTemplate(
        input_variables=["coin", "time_frame", "high_price", "low_price", "close_price", "volume", "risk_tolerance", "minimum_profit", "indicator_data"],
        template=template
    )

    # generate llm chain
    chain = LLMChain(llm=llm, prompt=prompt)

    # Run the chain only specifying the input variable.
    result = chain.run({
        'coin': coin,
        'time_frame': time_frame,
        'high_price': high_price,
        'low_price':  low_price,
        'close_price': close_price,
        'volume': volume,
        'risk_tolerance': risk_tolerance,
        'minimum_profit': minimum_profit,
        'indicator_data': indicator_data
        })
    
    return result
```
